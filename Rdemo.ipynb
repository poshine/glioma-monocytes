#### 2. Calculate the risk score using the model；

```r
### valid.pred
model.GLMnet.valid.pred <- lapply(model.GLMnet.mod, function(x){
  Prediction.Probs <- predict(x, newdata = t(test.exp[colnames(x$trainingData)[-ncol(x$trainingData)],]), type = "prob") %>% data.frame
  Prediction.Probs = cbind(test.ifo, Prediction.Probs)
  
})
```

```r
##
model.GLMnet.valid.pred.case = do.call(cbind, lapply(model.GLMnet.valid.pred, function(x) x$Case))
model.GLMnet.valid.pred.score = apply(model.GLMnet.valid.pred.case, 1, median)
```

```r
test.ifo$model.GLMnet.valid.pred.score = model.GLMnet.valid.pred.score
```
